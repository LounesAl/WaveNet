```python
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt 
import numpy as np
```

# MFFC

```python
audio_file='p225_001_mic1.flac'
#please call stella
audio_file='demo.wav'
```

```python
ipd.Audio(audio_file)
```

```python
signal,sr=librosa.load(audio_file)
```

```python
signal.shape
sr
```

```python
mfcc=librosa.feature.mfcc(signal,n_mfcc=13,sr=sr)
```

```python
mfcc.shape
```

```python
plt.figure(figsize=(25,10))
librosa.display.specshow(mfcc,x_axis='time',sr=sr)
plt.colorbar(format="%+2f")
plt.show()
```

```python
#calculate delata and delta2 MFFCS
delta_mfccs=librosa.feature.delta(mfcc)
delata2_mfccs=librosa.feature.delta(mfcc,order=2)
```

```python
delta_mfccs.shape
```

```python
plt.figure(figsize=(25,10))
librosa.display.specshow(delta_mfccs,x_axis='time',sr=sr)
plt.colorbar(format="%+2f")
plt.title('derivative of mfccs')
plt.show()
```

```python
plt.figure(figsize=(25,10))
librosa.display.specshow(delata2_mfccs,x_axis='time',sr=sr)
plt.colorbar(format="%+2f")
plt.show()
```

# Mel filter banks

```python

filter_banks = librosa.filters.mel(n_fft=2048, sr=22050, n_mels=10)
```

```python

filter_banks.shape
```

```python

plt.figure(figsize=(25, 10))
librosa.display.specshow(filter_banks, 
                         sr=sr, 
                         x_axis="linear")
plt.colorbar(format="%+2.f")
plt.show()
```

# Extracting Mel Spectrogram

```python
mel_spectrogram = librosa.feature.melspectrogram(signal, sr=sr, n_fft=2048, hop_length=512, n_mels=10)
```

```python

log_mel_spectrogram = librosa.power_to_db(mel_spectrogram)
```

```python
log_mel_spectrogram.shape
```

```python

plt.figure(figsize=(25, 10))
librosa.display.specshow(log_mel_spectrogram, 
                         x_axis="time",
                         y_axis="mel", 
                         sr=sr)
plt.colorbar(format="%+2.f")
plt.show()
```
